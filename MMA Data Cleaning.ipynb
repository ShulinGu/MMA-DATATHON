```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('dataset_Fall 2021.csv')
pd.set_option('display.max_columns', None)
df['F30_UPB'] = df['F30_UPB'].fillna(0)
df['F60_UPB'] = df['F60_UPB'].fillna(0)
df['F90_UPB'] = df['F90_UPB'].fillna(0)
df['F120_UPB'] = df['F120_UPB'].fillna(0)
df['F180_UPB'] = df['F180_UPB'].fillna(0)
df.head(10)
```

```python
dups = df.pivot_table(index = ['LAST_STAT'], aggfunc ='size')
dups
```

```python
POM = df[df['LAST_STAT'] == 'P']
POM['MI'] = POM['orig_rt']/1200
POM['MP'] = POM['orig_amt']*POM['MI']*(((1+POM['MI'])**POM['orig_trm'])/((1+POM['MI'])**POM['orig_trm']-1))

# Calculate the month
POM[["year", "month", "day"]] = POM["LAST_DTE"].str.split("-", expand = True)
POM[["YEAR", "MONTH", "DAY"]] = POM["FRST_DTE"].str.split("-", expand = True)
POM["MONTH"] = pd.to_numeric(POM["MONTH"])
POM["month"] = pd.to_numeric(POM["month"])
POM["YEAR"] = pd.to_numeric(POM["YEAR"])
POM["year"] = pd.to_numeric(POM["year"])
POM['Month'] = (POM['year']-POM['YEAR'])*12+POM['month']-POM['MONTH']
POM.drop(['MONTH','DAY','YEAR','month','day','year'], axis=1, inplace=True)

POM['Amount Received'] = POM['MP']*POM['Month']+POM['LAST_UPB']
POM['Expected Amount Receivable'] = POM['MP']*POM['orig_trm']
POM.head()
```

```python
MIP = df[df['LAST_STAT'] == 'C']
MIP['MI'] = MIP['orig_rt']/1200
MIP['MP'] = MIP['orig_amt']*MIP['MI']*(((1+MIP['MI'])**MIP['orig_trm'])/((1+MIP['MI'])**MIP['orig_trm']-1))

# Calculate the month
MIP[["year", "month", "day"]] = MIP["LAST_DTE"].str.split("-", expand = True)
MIP[["YEAR", "MONTH", "DAY"]] = MIP["FRST_DTE"].str.split("-", expand = True)
MIP["MONTH"] = pd.to_numeric(MIP["MONTH"])
MIP["month"] = pd.to_numeric(MIP["month"])
MIP["YEAR"] = pd.to_numeric(MIP["YEAR"])
MIP["year"] = pd.to_numeric(MIP["year"])
MIP['Month'] = (MIP['year']-MIP['YEAR'])*12+MIP['month']-MIP['MONTH']
MIP.drop(['MONTH','DAY','YEAR','month','day','year'], axis=1, inplace=True)

MIP['Amount Received'] = MIP['MP']*MIP['Month']
MIP['Expected Amount Receivable'] = MIP['MP']*MIP['orig_trm']
MIP.head()
```

```python
DEL_ALL = df[df['LAST_STAT'] == 'D']
DEL_ALL['MI'] = DEL_ALL['orig_rt']/1200
DEL_ALL['MP'] = DEL_ALL['orig_amt']*DEL_ALL['MI']*(((1+DEL_ALL['MI'])**DEL_ALL['orig_trm'])/((1+DEL_ALL['MI'])**DEL_ALL['orig_trm']-1))
DEL_ALL[["YEAR", "MONTH", "DAY"]] = DEL_ALL["FRST_DTE"].str.split("-", expand = True)
DEL_ALL["MONTH"] = pd.to_numeric(DEL_ALL["MONTH"])
DEL_ALL["YEAR"] = pd.to_numeric(DEL_ALL["YEAR"])

DEL_1 = DEL_ALL.loc[(DEL_ALL['F180_UPB'] > 0)]
DEL_2 = DEL_ALL.loc[(DEL_ALL['F180_UPB'] == 0) & (DEL_ALL['F120_UPB']> 0)]
DEL_3 = DEL_ALL.loc[(DEL_ALL['F120_UPB'] == 0) & (DEL_ALL['F90_UPB']> 0)]
DEL_4 = DEL_ALL.loc[(DEL_ALL['F90_UPB'] == 0) & (DEL_ALL['F60_UPB']> 0)]
DEL_5 = DEL_ALL.loc[(DEL_ALL['F60_UPB'] == 0) & (DEL_ALL['F30_UPB']> 0)]


DEL_1[["year", "month", "day"]] = DEL_1["F180_DTE"].str.split("-", expand = True)
DEL_1["month"] = pd.to_numeric(DEL_1["month"])
DEL_1["year"] = pd.to_numeric(DEL_1["year"])
DEL_1['Month'] = (DEL_1['year']-DEL_1['YEAR'])*12+DEL_1['month']-DEL_1['MONTH']
DEL_1['Amount Received'] = DEL_1['MP']*DEL_1['Month']
DEL_1.drop(['MONTH','DAY','YEAR','month','day','year'], axis=1, inplace=True)
DEL_1['Expected Amount Receivable'] = DEL_1['MP']*DEL_1['orig_trm']

DEL_2[["year", "month", "day"]] = DEL_2["F120_DTE"].str.split("-", expand = True)
DEL_2["month"] = pd.to_numeric(DEL_2["month"])
DEL_2["year"] = pd.to_numeric(DEL_2["year"])
DEL_2['Month'] = (DEL_2['year']-DEL_2['YEAR'])*12+DEL_2['month']-DEL_2['MONTH']
DEL_2['Amount Received'] = DEL_2['MP']*DEL_2['Month']
DEL_2.drop(['MONTH','DAY','YEAR','month','day','year'], axis=1, inplace=True)
DEL_2['Expected Amount Receivable'] = DEL_2['MP']*DEL_2['orig_trm']

DEL_3[["year", "month", "day"]] = DEL_3["F90_DTE"].str.split("-", expand = True)
DEL_3["month"] = pd.to_numeric(DEL_3["month"])
DEL_3["year"] = pd.to_numeric(DEL_3["year"])
DEL_3['Month'] = (DEL_3['year']-DEL_3['YEAR'])*12+DEL_3['month']-DEL_3['MONTH']
DEL_3['Amount Received'] = DEL_3['MP']*DEL_3['Month']
DEL_3.drop(['MONTH','DAY','YEAR','month','day','year'], axis=1, inplace=True)
DEL_3['Expected Amount Receivable'] = DEL_3['MP']*DEL_3['orig_trm']

DEL_4[["year", "month", "day"]] = DEL_4["F60_DTE"].str.split("-", expand = True)
DEL_4["month"] = pd.to_numeric(DEL_4["month"])
DEL_4["year"] = pd.to_numeric(DEL_4["year"])
DEL_4['Month'] = (DEL_4['year']-DEL_4['YEAR'])*12+DEL_4['month']-DEL_4['MONTH']
DEL_4['Amount Received'] = DEL_4['MP']*DEL_4['Month']
DEL_4.drop(['MONTH','DAY','YEAR','month','day','year'], axis=1, inplace=True)
DEL_4['Expected Amount Receivable'] = DEL_4['MP']*DEL_4['orig_trm']

DEL_5[["year", "month", "day"]] = DEL_5["F30_DTE"].str.split("-", expand = True)
DEL_5["month"] = pd.to_numeric(DEL_5["month"])
DEL_5["year"] = pd.to_numeric(DEL_5["year"])
DEL_5['Month'] = (DEL_5['year']-DEL_5['YEAR'])*12+DEL_5['month']-DEL_5['MONTH']
DEL_5['Amount Received'] = DEL_5['MP']*DEL_5['Month']
DEL_5.drop(['MONTH','DAY','YEAR','month','day','year'], axis=1, inplace=True)
DEL_5['Expected Amount Receivable'] = DEL_5['MP']*DEL_5['orig_trm']

MERGE = [DEL_1, DEL_2, DEL_3, DEL_4, DEL_5]
DEL = pd.concat(MERGE)
DEL.head(5)
#dup = DEL.pivot_table(index = ['LAST_STAT'], aggfunc ='size')
#dup
```

```python
pdList = [POM, MIP, DEL]
merge_1 = pd.concat(pdList)
merge_1.drop(['Month'], axis=1, inplace=True)
merge_1
```

```python
def func(x):
    if x <= 450:
        return "0-450"
    elif x<= 490:
        return "450-490"
    elif x <= 530:
        return "490-530"
    elif x <= 570:
        return "530-570"
    elif x <= 610:
        return "570-610"
    elif x <= 650:
        return "610-650"
    elif x <= 690:
        return "650-690"
    elif x <= 730:
        return "690-730"
    elif x <= 770:
        return "730-770"
    elif x <= 810:
        return "770-810"
    elif x <= 850:
        return "810-850"
    elif x > 850:
        return ">850"
    else:
        return 'NaN'
Final_Data['cscore_B'] = Final_Data['CSCORE_B'].apply(func)
Final_Data['cscore_C'] = Final_Data['CSCORE_C'].apply(func)
Final_Data
```
